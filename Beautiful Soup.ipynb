#### Import Libraries


```python
import requests
from bs4 import BeautifulSoup as bs
import time
import pandas as pd
```

### Scraping Pakwheels


#### Total pages = 473 (from site)


```python
carName = []
Price = []
driven = []
hP = []
mode = []
fuel = []
location = []
model = []
final = pd.DataFrame()
StartOverall = time.strftime("%H:%M:%S", time.gmtime())

for i in range(1,474):
    url = f"https://www.pakwheels.com/used-cars/search/-/ct_lahore/?page={i}"
    r = requests.get(url)
    soupObject = bs(r.content, 'html.parser')

    for elem in soupObject.find_all("h3"):
        if "View" not in elem.text.strip() and "?" not in elem.text.strip():
            carName.append(elem.text.strip()[:len(elem.text.strip())-8])

    for elem in soupObject.find_all("div", class_="price-details"):
        split_result = elem.text.strip()
        if len(split_result) > 1:
            Price.append(split_result)
        else:
            Price.append("nan")

    for elem in soupObject.find_all("div", class_="col-md-12 grid-date"):
        specs = elem.text.strip().split()
        if len(specs) >= 8:
            loc = specs[0]
            mdl = specs[1]
            drv = specs[2]
            fl = specs[4]
            hp = specs[5]
            mo = specs[7]

            location.append(loc)
            model.append(mdl)
            driven.append(drv)
            fuel.append(fl)
            hP.append(hp)
            mode.append(mo)
        else:
            location.append("nan")
            model.append("nan")
            driven.append("nan")
            fuel.append("nan")
            hP.append("nan")
            mode.append("nan")

min_length = min(len(carName), len(model), len(mode), len(driven), len(fuel), len(hP), len(location), len(Price))

# Create the dictionary with consistent lengths
data = {"Car": carName[:min_length],
        "Model": model[:min_length],
        "Mode": mode[:min_length],
        "Driven (km)": driven[:min_length],
        "Fuel": fuel[:min_length],
        "Horse Power": hP[:min_length],
        "Location": location[:min_length],
        "Price (Rs : lacs)": Price[:min_length]}

# Concatenate data to the existing DataFrame
final = pd.concat([final, pd.DataFrame(data)], ignore_index=True)

print("Time taken:",StartOverall," ", time.strftime("%H:%M:%S", time.gmtime()))
```
